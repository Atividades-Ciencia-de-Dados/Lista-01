````python
# Função para verificar valores incorretos, ou seja, do tipo errado
def verificar_valores_incorretos(df):
    tipos_esperados = {
        'date_message': 'datetime',
        'id_member_anonymous': 'str',
        'id_group_anonymous': 'str',
        'media': 'str',
        'media_type': 'str',
        'media_url': 'str',
        'has_media': 'bool',
        'has_media_url': 'bool',
        'trava_zap': 'bool',
        'text_content_anonymous': 'str',
        'dataset_info_id': 'int',
        'date_system': 'datetime',
        'score_sentiment': 'float',
        'score_misinformation': 'float',
        'id_message': 'int',
        'message_type': 'str',
        'messenger': 'str',
        'media_name': 'str',
        'media_md5': 'str'
    }

    problemas = []

    for coluna, tipo_esperado in tipos_esperados.items():
        if coluna in df.columns:
            for linha, valor in df[coluna].items():
                if pd.isna(valor):
                    continue  # pula valores nulos

                if tipo_esperado == 'datetime':
                    try:
                        pd.to_datetime(valor)
                    except Exception:
                        problemas.append((linha, coluna, valor, 'datetime'))

                elif tipo_esperado == 'float':
                    try:
                        float(valor)
                    except Exception:
                        problemas.append((linha, coluna, valor, 'float'))

                elif tipo_esperado == 'int':
                    try:
                        # Aqui checamos se é realmente um número inteiro
                        if not float(valor).is_integer():
                            problemas.append((linha, coluna, valor, 'int'))
                    except Exception:
                        problemas.append((linha, coluna, valor, 'int'))

                elif tipo_esperado == 'bool':
                    if valor not in [True, False]:
                        problemas.append((linha, coluna, valor, 'bool'))

                elif tipo_esperado == 'str':
                    if not isinstance(valor, str):
                        problemas.append((linha, coluna, valor, 'str'))

    return problemas
````

````python
# Verificar valores incorretos no DataFrame
problemas = verificar_valores_incorretos(df)
````

````python
# Criar DataFrame de problemas
df_problemas = pd.DataFrame(problemas, columns=['linha', 'coluna', 'valor_encontrado', 'tipo_esperado'])

# Mostrar
df_problemas
````

Como podemos ver acima, **não foram encontrados valores com tipos de dados incorretos**, de acordo com os tipos que defini previamente.

##k) Crie uma coluna chamada “sentiment” contendo os valores: -1 para textos negativos, 0 para textos neutros e 1 para textos positivos.

````python
from textblob import TextBlob

def analyze_sentiment(text):
    if pd.isna(text):  # Trata valores nulos como neutros
        return 0
    analysis = TextBlob(str(text))
    polarity = analysis.sentiment.polarity
    if polarity >= 0.05:
        return 1  # Positivo
    elif polarity <= -0.05:
        return -1  # Negativo
    else:
        return 0  # Neutro

df['sentiment'] = df['text_content_anonymous'].apply(analyze_sentiment)
````

Como foi possível observar, apenas 16 linhas estão marcadas como `True` na coluna `trava_zap`. No entanto, todas possuem valor `NaN` na coluna `text_content_anonymous`, o que impossibilita a análise de padrões diretamente no dataset.

Diante disso, foram consideradas duas alternativas:

- **Opção 1**: Remover as 16 linhas, já que o conteúdo está ausente (`NaN`), o que considero mais adequado neste caso.

- **Opção 2**: Aplicar expressões regulares (Regex) para identificar possíveis trava-zaps por meio de padrões como **muitas repetição de letras, emojis, palavras e simbolos especiais**, isso seguindo a ideia comum do WhatsApp, considerando o telegram outros que geram riscos são links encurtados, ofertas e promoções. No entanto, esse método é propenso a erros, pois nem toda mensagem com essas características configura um trava-zap.

Para a **Opção 1**, que consiste em remover as 16 linhas, basta executar o código abaixo:

````python
# Mantém apenas as linhas onde 'trava_zap' é False (ou seja, remove os "trava-zaps")
df_sem_trava_zaps = df[df['trava_zap'] == False]
````

Para a **Opção 2**, em que foram identificados possíveis trava-zaps utilizando Regex, basta executar o código abaixo:

````python
import re

# Regex para trava-zaps no Telegram
REGEX_TRAVA_ZAP = re.compile(
    r'(?:'                                     # Grupo de alternância não-capturante
    r'(?:\b(\w+)\b(?:\s+\1\b){3,})|'            # Mesma palavra repetida 4+ vezes
    r'[a-zA-Z]{100,}|'                          # Sequência enorme de letras
    r'([\U0001F600-\U0001F64F]){10,}|'           # Muitos emojis em Sequencia
    r'([^\w\s]){10,}'                           # Muitos símbolos especiais
    r')',
    flags=re.IGNORECASE
)

def detectar_trava_zaps(df_temp):
    """Detecta mensagens suspeitas sem modificar o DataFrame"""
    return df_temp[
        df_temp['text_content_anonymous']
        .str.contains(REGEX_TRAVA_ZAP, na=False)
    ]

# Cria uma cópia de df
df_analise = df.copy()

# Aplica a função na cópia
trava_zaps = detectar_trava_zaps(df_analise)

# Análise dos resultados
print(f"Mensagens suspeitas detectadas: {len(trava_zaps)}")
print(trava_zaps['text_content_anonymous'])
````
